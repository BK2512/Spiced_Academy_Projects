```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import KNNImputer
from sklearn.decomposition import NMF
import pickle
```

```python
ratings = pd.read_csv('ratings.csv')
ratings
```

```python
movies=pd.read_csv('movies.csv')
movies
```

```python
df=pd.merge(ratings,movies,on='movieId')
```

```python
df
```

```python
df.drop(columns=["timestamp"],inplace=True)
```

```python
df
```

```python
df.set_index('userId')
```

```python
df
```

```python
df[df['title']=='\'Hellboy\': The Seeds of Creation (2004)']
```

```python
df2=pd.pivot_table(df,index='userId',columns='title',values='rating')
```

```python
df2
```

```python
df2.isna().sum()
```

```python
imputer=KNNImputer(n_neighbors=2)
```

```python
dfrecs=pd.DataFrame(imputer.fit_transform(df2), 
                       index = df2.index, 
                       columns = df2.columns)
```

```python
dfrecs
```

```python
#dfrecs.to_csv('Movie_recommender.csv', index=True)
```

###### Create a model

```python
model=NMF(n_components=20,init='random',random_state=42,max_iter=10000)
model.fit(dfrecs)
```

```python
model.components_
```

```python
Q=pd.DataFrame(model.components_,
              columns=dfrecs.columns,
              index=['feature1','feature2','feature3','feature4','feature5','feature6','feature7','feature8','feature9','feature10',
                    'feature11','feature12','feature13','feature14','feature15','feature16','feature17','feature18','feature19','feature20'])
```

```python
Q
```

```python
P=pd.DataFrame(model.transform(dfrecs),index=dfrecs.index,
                 columns=['feature1','feature2','feature3','feature4','feature5','feature6','feature7','feature8','feature9','feature10',
                         'feature11','feature12','feature13','feature14','feature15','feature16','feature17','feature18','feature19','feature20'])
                 
```

```python
P
```

```python
recom_df=pd.DataFrame(np.dot(P,Q),index=dfrecs.index,columns=dfrecs.columns)
```

```python
recom_df
```

```python
a=dfrecs-recom_df
a
```

```python
model.reconstruction_err_
```
