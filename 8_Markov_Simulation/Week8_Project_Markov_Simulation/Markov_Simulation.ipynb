###### create a crosstable to see the Probabilities

```python
crosstable=pd.crosstab(df['location'], df['after'], normalize=0)
crosstable
```

###### Creating a new class: Customer

```python
class Customer:
    """
    a single customer that moves through the supermarket
    in a MCMC simulation
    """
    def __init__(self,name,state,budget=100):
        self.name=name
        self.state=state
        self.budget=budget
        
    def is_active(self):
        """"
        Returns True if customer if customer is still in store and not reached the checkout
        """
        if self.state!='checkout':
            return True
    
    def next_state(self):
        """
        Propagates the customer to the next state.
        Returns nothing.
        """
        self.state=np.random.choice(['checkout','dairy','drinks','fruit','spices'],p=crosstable.loc[self.state])
        
    def __repr__(self):
        return f'<Customer {self.name} is in {self.state}>'
```

###### writing a while-loop, to see which stations BK/LB are visiting before checking out

```python
BK=Customer('BK','dairy')
while BK.next_state()!='checkout':
    print(BK)
    if BK.state=='checkout':
        break
```

```python
LB=Customer('LB','spices')
while LB.next_state()!='checkout':
    print(LB)
    if LB.state=='checkout':
        break
```
